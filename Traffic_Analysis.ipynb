```python
import pandas as pd
import numpy as np


df= pd.read_csv(r'/Users/jakefischer/DivvyScrape/div_data4.csv')
df

#This will measure the traffic in and out of the stations
b= df.iloc[0:len(df.index),3:len(df.columns)].to_numpy()
traffic = np.zeros((b.shape[0],b.shape[1]-1))
for i in range(traffic.shape[0]):
    for j in range(traffic.shape[1]):
        traffic[i][j] = b[i][j+1]-b[i][j]


#absolute traffic
traffic = np.abs(traffic)
traffics = np.zeros(traffic.shape[0])

for k in range(traffic.shape[0]):
    traffics[k] = sum(traffic[k][:])/len(traffic[k][:])
    
    
df['absolute traffic'] = traffics
```

```python
df = df.sort_values(by=['absolute traffic'], ascending=False)
df.to_csv(r'/Users/jakefischer/DivvyScrape/traffic_div_data4.csv')
```

```python
#This measures the traffic leaving each of the stations


c= df.iloc[0:len(df.index),3:len(df.columns)].to_numpy()
thrutraffic = np.zeros((c.shape[0],c.shape[1]-1))
for i in range(thrutraffic.shape[0]):
    for j in range(thrutraffic.shape[1]):
        thrutraffic[i][j] = c[i][j+1]-c[i][j]


#traffic leaving
traffic =filter(lambda x : x > 0,thrutraffic)

thrutraffics = np.zeros(thrutraffic.shape[0])

for k in range(thrutraffic.shape[0]):
    thrutraffics[k] = sum(thrutraffic[k][:])/len(thrutraffic[k][:])
    
    
df['traffic leaving'] = thrutraffics
```

```python
thru = df.sort_values(by=['traffic leaving'], ascending=False)


lat = thru['lattitude'].to_numpy()
long = thru['longitude'].to_numpy()
map = folium.Map(location=[lat.mean(), long.mean()], zoom_start=14, control_scale=True)


k = 100
for i in range(k):
   folium.Marker(location=[lat[i], long[i]], popup = thru.iloc[i][0], icon=folium.Icon(color='red')).add_to(map)

for i in range(k, 740):
    folium.Marker(location=[lat[i], long[i]], popup = thru.iloc[i][0], icon=folium.Icon(color='blue')).add_to(map)

map
```
