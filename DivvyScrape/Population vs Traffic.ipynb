```python
import pandas as pd
import numpy as np

locations = pd.read_csv(r'/Users/jakefischer/Divvy_repo/DivvyScrape/tract_and_location1.csv')
tracpop = pd.read_csv(r'/Users/jakefischer/Divvy_repo/DivvyScrape/tract_populations.csv')
traffic = pd.read_csv(r'/Users/jakefischer/Divvy_repo/DivvyScrape/traffic_div_data4.csv')
locations
```

```python
#review this!!!

lats = []
longs = []
for i in list(tracpop.index):
    if(i in list(locations.index)):
        lats.append(locations.loc[i, "latitude"])
        longs.append(locations.loc[i, "longitude"])
    else:
        lats.append(float("nan"))
        longs.append(float("nan"))
    
tracpop["latitudes"] = lats
tracpop["longitudes"] = longs
```

```python
for k in range(len(tracpop.index)):
    if(np.isnan(tracpop.loc[(tracpop.index)[k],"latitudes"])):
        tracpop = tracpop.drop(tracpop.index[k], axis = 0)
```

```python
#second attempt at cell above
#tracpop 800x5
# traffic.shape(745,125)

stationtracts = []

#assocate each station with a tract
for i in list(traffic.index):
    laty = np.full(shape = len(tracpop.index), fill_value = traffic.loc[i,"lattitude"])
    latdif = np.abs(laty-list(tracpop['latitudes']))
    
    
    
    longy = np.full(shape = len(tracpop.index), fill_value = traffic.loc[i,"longitude"])
    longdif = np.abs(longy-list(tracpop['longitudes']))
    
    #distances from station to every tract
    d = np.sqrt((np.square(latdif)+np.square(longdif)))
    #d = np.sqrt(longdif*longdif+latdif*latdif)
    #this is the index of the closest tract to the station from tracpop
    theone = np.argmin(d)
    stationtracts.append(theone)
   
thetract = []
for i in stationtracts:
    thetract.append(tracpop.loc[i,"tracts"])
    

traffic["tract"] = thetract


    
```
